```python
final_census_data = pd.read_csv("final_census_data.csv")
df = final_census_data
```

```python
#Create a bar graph to show Population count from 2011 to 2018
df.pivot(index='Year',columns='City', values='Population').plot(kind="bar")
plt.title("CLT vs SF Population")
plt.xlabel('Years')
plt.ylabel('Population Count')

plt.savefig("Population Bar Graph")
```

```python
#Create a line graph to show Population count from 2011 to 2018
from scipy.stats import linregress
x_values = final_census_data["Year"]
y_values = final_census_data["Population"]
slope, intercept, rvalue, pvalue, stderr = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2015, 750000),fontsize=15,color="red")
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('CLT vs SF Population')
plt.show()
plt.savefig("Population Line Graph")
```
